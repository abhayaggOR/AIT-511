```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("train.csv")
```

```python
df['BMI'] = df['Weight'] / (df['Height'] ** 2)
```

```python
# Count occurrences of each transport type
mtrans_counts = df['MTRANS'].value_counts()

# Display the counts
print(mtrans_counts)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Filter only these four transport types
selected_trans = [ 'Walking', 'Motorbike', 'Bike']
subset = df[df['MTRANS'].isin(selected_trans)]

# Create a cross-tabulation of counts
plot_data = pd.crosstab(subset['MTRANS'], subset['WeightCategory'])

# Plot the grouped bar chart
plot_data.plot(kind='bar', figsize=(10,6), edgecolor='black')

plt.title('Weight Category Distribution by Transportation Mode')
plt.xlabel('Transportation Mode (MTRANS)')
plt.ylabel('Number of Individuals')
plt.legend(title='Weight Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Filter only the 4 transport types
selected_trans = ['Public_Transportation','Automobile', 'Walking', 'Motorbike', 'Bike']
subset = df[df['MTRANS'].isin(selected_trans)]

# Create crosstab (count of WeightCategory per MTRANS)
ratio_table = pd.crosstab(subset['MTRANS'], subset['WeightCategory'], normalize='index') * 100

# Round to 2 decimals for clarity
ratio_table = ratio_table.round(2)

# Display the ratio table
print("Percentage distribution of Weight Categories within each Transportation Mode:")
print(ratio_table)
```
