```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("train_combined.csv")
```

```python
df['BMI'] = df['Weight'] / (df['Height'] ** 2)
```

```python
df['Age'] = df['Age'].round().astype(int)
```

```python
order = [
    'Insufficient_Weight',
    'Normal_Weight',
    'Overweight_Level_I',
    'Overweight_Level_II',
    'Obesity_Type_I',
    'Obesity_Type_II',
    'Obesity_Type_III'
]
```

```python
# Violin plot
plt.figure(figsize=(10,6))
sns.violinplot(data=df, x='WeightCategory', y='BMI', order=order, palette='viridis', inner='quartile', cut=0)

plt.title('BMI Distribution Across Weight Categories')
plt.xlabel('Weight Category')
plt.ylabel('BMI')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```
