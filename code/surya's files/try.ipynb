```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
from xgboost import XGBClassifier   # or any classifier youâ€™ve chosen
from sklearn.pipeline import Pipeline

# 1ï¸âƒ£ Read your train data
train_df = pd.read_csv("train.csv")  # replace with your dataset file if needed

# 2ï¸âƒ£ Separate features and target
X = train_df.drop("WeightCategory", axis=1)
y = train_df["WeightCategory"]

# 3ï¸âƒ£ Define categorical columns
categorical_features = [
    "Gender",
    "family_history_with_overweight",
    "FAVC",
    "CAEC",
    "SMOKE",
    "SCC",
    "CALC",
    "MTRANS"
]

# 4ï¸âƒ£ Define the preprocessing transformer
one_hot = OneHotEncoder(handle_unknown="ignore", sparse_output=False)
transformer = ColumnTransformer(
    transformers=[("one_hot", one_hot, categorical_features)],
    remainder="passthrough"
)

# 5ï¸âƒ£ Split train/test for fitting SMOTE and model
X_train, X_val, y_train, y_val = train_test_split(
    X, y, stratify=y, test_size=0.2, random_state=8
)

# 6ï¸âƒ£ Transform training features
X_train_transformed = transformer.fit_transform(X_train)
X_val_transformed = transformer.transform(X_val)

# 7ï¸âƒ£ Apply SMOTE
smote = SMOTE(random_state=8)
X_train_res, y_train_res = smote.fit_resample(X_train_transformed, y_train)

# 8ï¸âƒ£ Train model (example: XGBoost)
model = XGBClassifier(
    random_state=8,
    n_estimators=300,
    learning_rate=0.1,
    max_depth=6,
    subsample=0.8,
    colsample_bytree=0.8
)
model.fit(X_train_res, y_train_res)

# âœ… === Inference stage ===

# 9ï¸âƒ£ Load test data
test_df = pd.read_csv("test.csv")

# 10ï¸âƒ£ Apply same preprocessing
X_test_transformed = transformer.transform(test_df)

# 11ï¸âƒ£ Predict WeightCategory
preds = model.predict(X_test_transformed)

# 12ï¸âƒ£ Prepare submission DataFrame
submission = pd.DataFrame({
    "id": test_df["id"],
    "WeightCategory": preds
})

# 13ï¸âƒ£ Save to CSV
submission.to_csv("sample_submission.csv", index=False)

print("âœ… sample_submission.csv saved!")
print(submission.head())
```
